```python
import numpy as np
import pandas as pd
source = "HDR 2007-2008 Table 14.xlsx"

df = pd.read_excel(source)
df.head(12)
```

```python
df = pd.read_excel(source, header = 4)
#suppression des colonnes sans valeurs
df = df.dropna(how = "all", axis = 1)
#suppression des lignes sans valeurs
df = df.dropna(how = "all", axis = 0).reset_index(drop = True)
df.head(20)
```

```python
columns = []
for c in df.columns:
    columns.append([str(v) for v in df[c][:4] if pd.notnull(v)])
columns
```

```python
columns = ["HDI rank",
          "Country",
          "GDP - US$ billions 2005",
          "References 1",
          "GDP - PPP US $ billions 2005",
          "References 2",
          "GDP per capita - US$ 2005",
          "References 3",
          "GDP per capita - 2005 PPP US$ 2005",
          "References 4",
          "GDP per capita - Annual growth rate(%) 1975-2005",
          "References 5",
          "GDP per capita - Annual growth rate(%) 1990-2005",
          "References 6",
          "GDP per capita - highest value during 1975-2005 2005 PPP US$",
          "References 7",
          "GDP per capita - year of highest value",
          "Average annual change in consumer price index(%) - 1990-2005",
          "Average annual change in consumer price index(%) - 2004-2005"]
```

```python
#réinitialiser les colonnes des cadres de données
df.columns = columns
```

```python
df = df.drop(np.arange(4), axis=0).reset_index(drop=True)
df.head()
```

```python
df.tail(30)
```

```python
df = df.drop(np.arange(180,len(df)), axis=0).reset_index(drop=True)
```

```python
hdi_categories = ["HIGH HUMAN DEVELOPMENT", "MEDIUM HUMAN DEVELOPMENT", "LOW HUMAN DEVELOPMENT"]
df = df.drop(df[df["HDI rank"].isin(hdi_categories)].index, axis=0).reset_index(drop=True)
df.loc[:, "HDI category"] = np.where(df["HDI rank"] > 155,
                                     "LOW HUMAN DEVELOPMENT",
                                     np.where(df["HDI rank"] > 70,
                                              "MEDIUM HUMAN DEVELOPMENT",
                                              "HIGH HUMAN DEVELOPMENT"
                                             )
                                    )
df.sort_values(by="HDI rank", inplace=True)

df.head()
```

```python
destination = "economic_performance.csv"
# `index=False` signifie que nous n'allons pas garder cette colonne d'index - elle sera recréée lorsque nous ouvrirons le fichier à l'avenir
df.to_csv(destination, index=False)
```
